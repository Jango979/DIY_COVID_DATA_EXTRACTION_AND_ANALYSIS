```python
import pandas as pd
```

```python
DATA=pd.read_csv('owid-covid-data.csv')



List=DATA['location']
print(List)
List=List.duplicated();List=[not i for i in List];
x=DATA.iloc[List,2]
print(x)


def EXTRACTION(CHOOSEN,CATEGORY):
    Extracted=[]
    for i in range(0,len(DATA['location'])):
        if CHOOSEN==DATA['location'][i]:
            Extracted.append(DATA[CATEGORY][i])
    return Extracted

    
```

```python
Countries=x.drop([25650],axis=0)
Categories=list(DATA.columns);
print(Categories)
for i in ['iso_code','continent','location']:
    Categories.remove(i)
print(Categories)
```

```python
DF1=pd.DataFrame({'' : []})


for i in Countries:
    for j in Categories:
        Extraction=EXTRACTION(i,j)
        DFTest=pd.DataFrame({str(i)+'_'+str(j):Extraction})
        DF1=pd.concat([DF1,DFTest],axis=1)
```

```python

DF1.to_csv('DATA.csv')
```

```python
import pandas as pd
America=pd.DataFrame({
    'Antigua and Barbuda':DF['Antigua and Barbuda_total_cases'],
    'Argentina':DF['Argentina_total_cases'],
    'Bahamas':DF['Bahamas_total_cases'],
    'Barbados':DF['Barbados_total_cases'],
    'Bolivia':DF['Bolivia_total_cases'],
    'Brazil':DF['Brazil_total_cases'],
    'Chile':DF['Chile_total_cases'],
    'Colombia':DF['Colombia_total_cases'],
    'Costa Rica':DF['Costa Rica_total_cases'],
    'Cuba':DF['Cuba_total_cases'],
    'Ecuador':DF['Ecuador_total_cases'],
    'El Salvador':DF['El Salvador_total_cases'],
    'Haiti':DF['Haiti_total_cases'],
    'Honduras':DF['Jamaica_total_cases'],
    'Mexico':DF['Mexico_total_cases'],
    'Nicaragua':DF['Nicaragua_total_cases'],
    'Panama':DF['Paraguay_total_cases'],
    'Peru':DF['Peru_total_cases'],
    'Uruguay':DF['Uruguay_total_cases'],
    'Venezuela':DF['Venezuela_total_cases'],
    'United States':DF['United States_total_cases'],
    'Canada':DF['Canada_total_cases']
})

America=America.replace(NaN,0)

AmericaFiltered={}
for Countries in America:
    
    ToAmericaDF=[]
    for OnlyData in range(0,len(America[Countries])):
        if America[Countries][OnlyData]!=0:
            ToAmericaDF.append(America[Countries][OnlyData])
    AmericaFiltered[Countries]=ToAmericaDF[len(ToAmericaDF)-1]
print(AmericaFiltered)


```
